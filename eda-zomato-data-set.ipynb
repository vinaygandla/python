```python
# how many of zomato the in the country
country_names=c.Country.value_counts().index
```

```python
# to get the data in a data frame reset_index is used 
d=c.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:"rating_count"})
```

```python
d
```

# observations....
1.when ever the rating is from 4.5 to 4.9 the food is really Excellent and with a proper delivery
2.when ever the rating is from 4.0 to 4.4 the food is very good and with a proper delivery 
3.when ever the rating is from 3.5 to 3.9 the food is good and with a proper delivery 
4.when ever the rating is from 2.5 to 3.4 the food is average and the also delivery may not be good 
5.when ever the rating is from 1.8 to 2.4 the food is poor and also delivery may not be good 
6.if the rating is 0.0 then we conclude that the customer had not given the rating
